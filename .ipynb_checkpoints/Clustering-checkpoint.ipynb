# Clustering
Comparing K-Means and DBSCAN clustering algorithms

```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_circles, make_blobs
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="darkgrid")
plt.style.use('fivethirtyeight')
```

## Create Datasets

```python
n_samples = 1500
X_circles, y_circles = make_circles(n_samples=n_samples, factor=.5, noise=0.05)
ax = sns.scatterplot(x=X_circles[:,0],y=X_circles[:,1],hue=y_circles)
```

```python
X_blobs, y_blobs = make_blobs(n_samples=n_samples, centers=3, n_features=2, random_state=21)
ax1 = sns.scatterplot(x=X_blobs[:,0],y=X_blobs[:,1],hue=y_blobs)
```

## K-Means

### Model from Scratch
Based on: https://pythonprogramming.net/k-means-from-scratch-machine-learning-tutorial/

### Model in Scikit-Learn

```python
from sklearn.cluster import KMeans

k_means_blobs = KMeans(n_clusters=3,random_state=21).fit(X_blobs)
k_means_blobs_plot = sns.scatterplot(x=X_blobs[:,0],y=X_blobs[:,1],hue=k_means_blobs.labels_)
```

```python
k_means_circles = KMeans(n_clusters=2,random_state=21).fit(X_circles)
k_means_circles_plot = sns.scatterplot(x=X_circles[:,0],y=X_circles[:,1],hue=k_means_circles.labels_)
```

## DBSCAN

### Model from Scratch
Based on: http://mccormickml.com/2016/11/08/dbscan-clustering/

### Model in Scikit-Learn

```python
from sklearn.cluster import DBSCAN

dbscan_blobs = DBSCAN().fit(X_blobs)
dbscan_blobs_plot = sns.scatterplot(x=X_blobs[:,0],y=X_blobs[:,1],hue=dbscan_blobs.labels_)
```

```python
## NOTE: DBSCAN is dependent on the value chosen for eps and min_smaples
## try with different values to see if the clustering returns expected results
dbscan_circles = DBSCAN(eps=.1).fit(X_circles)
dbscan_circles_plot = sns.scatterplot(x=X_circles[:,0],y=X_circles[:,1],hue=dbscan_circles.labels_)
```

```python
dbscan_circles = DBSCAN(eps=.4).fit(X_circles)
dbscan_circles_plot = sns.scatterplot(x=X_circles[:,0],y=X_circles[:,1],hue=dbscan_circles.labels_)
```
